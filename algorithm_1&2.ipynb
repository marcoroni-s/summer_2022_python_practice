```python
def algorithm_1(message):
    previous_character = ""
    shortened_message = ""
    vowel_removed = 0
    repeat_removed = 0

    for character in message:
        if character.lower() in "aeiou":  # looks for a vowel immediately and after a space 
            if previous_character == "" or previous_character == " ": 
                shortened_message = shortened_message + character  # Keeps vowel and makes it part of shortened message
            else:
                vowel_removed = vowel_removed + 1  # Adds to the counter if vowel isn't at the start of each word 
        elif character.lower() == previous_character.lower():
            repeat_removed = repeat_removed + 1  # Repeat checker 
        else:
            shortened_message = shortened_message + character  # Adds it if it isn't a repeat or a vowel
        
        previous_character = character  # Updater 
    
    characters_saved = len(message) - len(shortened_message) 
    
    return (vowel_removed, repeat_removed, shortened_message, characters_saved)


def algorithm_2(message):
    shortened_message = ""
    unique_characters = {""}
    characters_saved = 0

    for character in message:
        if character != " " and character.lower() not in unique_characters:
            unique_characters.add(character.lower())
            count = message.lower().count(character.lower())
            shortened_message = shortened_message + str(count) + character.lower()
            characters_saved = characters_saved + count - 1

    characters_saved = len(message) - len(shortened_message)

    return len(unique_characters), shortened_message, characters_saved




#ask for user input and calls both functions 
message = input("Type the message to be shortened: ").lower()
print("\n")

algorithm_1_result = algorithm_1(message)
print("Algorithm 1")
print("Vowels removed:", algorithm_1_result[0])
print("Repeats removed:", algorithm_1_result[1])
print("Algorithm 1 message:", algorithm_1_result[2])
print("Algorithm 1 characters saved:", algorithm_1_result[3])
print("\n")

algorithm_2_result = algorithm_2(message)
print("Algorithm 2")
print("Unique characters found:", algorithm_2_result[0])
print("Algorithm 2 message:", algorithm_2_result[1])
print("Algorithm 2 characters saved:", algorithm_2_result[2])

```
